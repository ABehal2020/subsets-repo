```python
import scipy.io
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import scipy.stats

# Load the provided .mat file to examine its contents
file_path = '../Data/01_Subset_Size_Match_Rate.mat'
data = scipy.io.loadmat(file_path)

data.keys()
```

```python
# Flatten the arrays for plotting
subset_sizes_spa_flat = data['subsetSizesSpa'].flatten()
subset_sizes_cue_flat = data['subsetSizesCue'].flatten()

# Create a DataFrame for seaborn plotting
df = pd.DataFrame({
    'Subset Size': np.concatenate((subset_sizes_spa_flat, subset_sizes_cue_flat)),
    'Task': ['Spatial']*len(subset_sizes_spa_flat) + ['Cue']*len(subset_sizes_cue_flat)
})

# Plotting the swarmplot with modified y-axis
plt.figure(figsize=(3, 5))
ax = sns.swarmplot(x='Task', y='Subset Size', data=df, size=10, color='black')  # Increase dot size
ax.set_ylim(bottom=0)  # Ensuring the y-axis starts at 0

# Remove top and right borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Customize layout
ax.set_xlabel('Task')
ax.set_ylabel('Proportion of Neurons per Ensemble in Predictive Subset')
plt.tight_layout()

# Save the figure
plt.savefig('Fig_1B.tiff', format='tiff')
plt.show()
```

```python
print("Subset Sizes Spa Stats: %f ± %f" % (np.mean(subset_sizes_spa_flat), scipy.stats.sem(subset_sizes_spa_flat, ddof=0)))
print("Subset Sizes Cue Stats: %f ± %f" % (np.mean(subset_sizes_cue_flat), scipy.stats.sem(subset_sizes_cue_flat, ddof=0)))
```

```python
# Use the overlap data for plotting
overlap_values = data['matchRateSpaCue'].flatten()

# Create a DataFrame for seaborn plotting for overlap
df_overlap = pd.DataFrame({
    'Overlap': overlap_values,
    'Ensemble': ['Ensemble'] * len(overlap_values)
})

# Plotting the swarmplot for overlap
plt.figure(figsize=(2, 5))
ax = sns.swarmplot(x='Ensemble', y='Overlap', data=df_overlap, size=10, color='black')
ax.set_xlabel('')
ax.set_ylabel('Proportion of Spatial and Cue Predictive Subset Overlap')

# Remove top and right borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Customize layout
plt.tight_layout()

# Save the figure
plt.savefig('Fig_1C.tiff', format='tiff')
plt.show()
```

```python
print("Overlap: %f ± %f" % (np.mean(overlap_values), scipy.stats.sem(overlap_values, ddof=0)))
```
